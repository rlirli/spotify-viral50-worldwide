# Spotify Viral 50 for every country

I needed a list with the Viral 50 and Top 50 playlists of all countries. 

Could not get them directly via the Spotify Web API categories. Instead I used the first result from Google Search queries such as "Spotify Viral 50 xxCountry".

This notebook searches for each of the 92 countries in which Spotify is available as of August 14th 2020. Extend the accompanying **spotify-countries.csv** to add more countries or use any other search term. The list of playlists is saved in **playlist_urls_viral_and_top.csv**. While some links return false, it works for most countries.

Skip to the last cell to recreate the final list of playlist urls.

```python
# retrieving the first result from Google Search
# thanks to Kevin Guan https://stackoverflow.com/a/33427649/9761761

import re
import requests
from bs4 import BeautifulSoup


country = "spain" # test country
goog_search = "https://www.google.com/search?q=spotify+viral+50+" + country

r = requests.get(goog_search)

soup = BeautifulSoup(r.text, 'html.parser')

goog_res_first = soup.find('div', {'class': 'kCrYT'})
goog_res_link = goog_res_first.findChildren("a" , recursive=False)

s = str(goog_res_link)

#filter out the "open.spotify.com/playlist/..." url
result = re.search('https://(.*)&amp;sa', s)
url_playlist = result.group(1)
print(country, ": ", url_playlist)
```

```python
# loop all countries save results in playlist_urls.csv

import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import csv

data = pd.read_csv("spotify-countries.csv", sep = ';', header = None)

for i, r in data.iterrows():
    country_iso = r[0]
    country = r[1]

    try:
        goog_search = "https://www.google.com/search?q=spotify+viral+50+" + country

        r = requests.get(goog_search)

        soup = BeautifulSoup(r.text, 'html.parser')

        goog_res_first = soup.find('div', {'class': 'kCrYT'})
        goog_res_link = goog_res_first.findChildren("a" , recursive=False)

        s = str(goog_res_link)
        result = re.search('https://(.*)&amp;sa', s)
        url_playlist = result.group(1)
        print(country, ": ", url_playlist)
        
        
        fields=[country_iso, country, url_playlist]
        with open(r"playlist_urls.csv", 'a') as f:
            writer = csv.writer(f)
            writer.writerow(fields)
    except:
        print("ERR", country)
        
        fields=[country_iso, country, "ERR"]
        with open(r"playlist_urls.csv", "a") as f:
            writer = csv.writer(f)
            writer.writerow(fields)
```

# Viral 50 and Top 50 combined

```python
# loop all countries save results in playlist_urls.csv
# result only saved in CSV. not printed in jupyter notebook

import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import csv

data = pd.read_csv("spotify-countries.csv", sep = ';', header = None)

for i, r in data.iterrows():
    country_iso = r[0]
    country = r[1]

    try:
        goog_search = "https://www.google.com/search?q=spotify+viral+50+" + country

        r = requests.get(goog_search)

        soup = BeautifulSoup(r.text, 'html.parser')

        goog_res_first = soup.find('div', {'class': 'kCrYT'})
        goog_res_link = goog_res_first.findChildren("a" , recursive=False)

        s = str(goog_res_link)
        result = re.search('https://(.*)&amp;sa', s)
        url_playlist_viral = result.group(1)
        if url_playlist_viral.rpartition("/")[0] == "open.spotify.com/playlist":
            spotifyID_viral = url_playlist_viral.rpartition("/")[2] # ugly TODO fix
        else:
            spotifyID_viral = "INCOMPATIBLE LINK"
    except:
        url_playlist_viral = "ERR"
        spotifyID_viral = "ERR"
    
    try:
        goog_search = "https://www.google.com/search?q=spotify+top+50+" + country

        r = requests.get(goog_search)

        soup = BeautifulSoup(r.text, 'html.parser')

        goog_res_first = soup.find('div', {'class': 'kCrYT'})
        goog_res_link = goog_res_first.findChildren("a" , recursive=False)

        s = str(goog_res_link)
        result = re.search('https://(.*)&amp;sa', s)
        url_playlist_top = result.group(1)
        if url_playlist_top.rpartition("/")[0] == "open.spotify.com/playlist":
            spotifyID_top = url_playlist_top.rpartition("/")[2]
        else:
            spotifyID_top = "INCOMPATIBLE LINK"
    except:
        url_playlist_top = "ERR"
        spotifyID_top = "ERR"
        
        
    fields=[country_iso, country, url_playlist_viral, spotifyID_viral, url_playlist_top, spotifyID_top]
    with open(r"playlist_urls_viral_and_top.csv", 'a') as f:
        writer = csv.writer(f)
        writer.writerow(fields)
```
